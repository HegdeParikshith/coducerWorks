```python
import fitz  # PyMuPDF
from docx import Document
from docx.shared import Pt
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.oxml import OxmlElement

def add_text_with_formatting(doc, text, style, font_size=12, color=None, alignment=None):
    run = doc.add_paragraph().add_run(text)
    
    if style:
        run.bold = 'bold' in style
        run.italic = 'italic' in style
        run.underline = 'underline' in style
    
    if color:
        run.font.color.rgb = color
    
    if alignment:
        doc.paragraphs[-1].alignment = alignment
    
    run.font.size = Pt(font_size)

def pdf_to_docx_with_formatting(pdf_path, docx_path):
    pdf_document = fitz.open(pdf_path)
    doc = Document()

    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        page_text = page.get_text("text")
        
        for block in page_text:
            text = block[4]  # Extract the text
            style = block[6]  # Extract the style attributes (e.g., "bold italic")
            font_size = block[7]  # Extract the font size (e.g., 12)
            color = block[8]  # Extract the font color
            alignment = block[9]  # Extract the alignment

            add_text_with_formatting(doc, text, style, font_size, color, alignment)

    doc.save(docx_path)

# Specify the paths for the input PDF and the output DOCX
pdf_file_path = 'Parikshitha_Hegde_Resume.pdf'
docx_output_path = 'output.docx'

# Call the function to convert the PDF and retain formatting in DOCX
pdf_to_docx_with_formatting(pdf_file_path, docx_output_path)
```

```python
import io
from PyPDF2 import PdfReader
from docx import Document
from docx.shared import Pt

def read_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PdfReader(file)
        text = ""
        for page_num in range(reader.pages):
            text += reader.getPage(page_num).extractText()
        return text

def write_docx(text, docx_path):
    document = Document()
    for line in text.split('\n'):
        document.add_paragraph(line)
    document.save(docx_path)

def convert_pdf_to_docx(pdf_path, docx_path):
    text = read_pdf(pdf_path)
    write_docx(text, docx_path)

# Example usage:
pdf_path = './Parikshitha_Hegde_Resume.pdf'
docx_path = 'output.docx'
convert_pdf_to_docx(pdf_path, docx_path)
```

```python
import subprocess

def pdf_to_html(input_pdf):
    try:
        # Run the pdf2htmlEX command using subprocess and capture the output
        result = subprocess.run(["pdf2htmlEX", input_pdf, "-"], stdout=subprocess.PIPE, text=True)
        
        if result.returncode == 0:
            # If the conversion was successful, return the HTML content as a string
            html_content = result.stdout
            return html_content
        else:
            print("Conversion failed")
            return None
    except Exception as e:
        print(f"Conversion failed: {e}")
        return None

if __name__ == "__main__":
    # Specify the input PDF file
    input_pdf = "./Parikshitha_Hegde_Resume.pdf"

    html_string = pdf_to_html(input_pdf)

    if html_string:
        # Print or use the HTML content as needed
        print(html_string)
```

```python
import docx
from bs4 import BeautifulSoup

def extract_styles(run):
    styles = ""
    if run.bold:
        styles += "font-weight: bold;"
    if run.italic:
        styles += "font-style: italic;"
    if run.underline:
        styles += "text-decoration: underline;"
    return styles

def docx_to_html_with_styles(input_docx, output_html):
    doc = docx.Document(input_docx)

    # Create an empty HTML string
    html = "<html><head><style> p { %s } </style></head><body>" % extract_styles(doc.paragraphs[0].runs[0])

    for paragraph in doc.paragraphs:
        # Add each paragraph as a <p> element in the HTML with styles
        styles = extract_styles(paragraph.runs[0])
        print(styles)
        html += f"<p style='{styles}'>{paragraph.text}</p>"

    html += "</body></html>"

    # Parse the HTML string using BeautifulSoup for prettifying
    soup = BeautifulSoup(html, 'html.parser')
    pretty_html = soup.prettify()

    # Save the pretty HTML to the output file
    with open(output_html, 'w', encoding='utf-8') as output_file:
        output_file.write(pretty_html)
   
input_docx = "./PdfToDocx2.docx"
output_html = "./output.html"
docx_to_html_with_styles(input_docx, output_html)
```

```python
import os
import json
from pdf2docx import Converter
from docx import Document
import pytesseract
from PIL import Image
from io import BytesIO
import docx2txt

def pdf_to_slatejs_json(pdf_path):
    # Convert PDF to DOCX using pdf2docx while retaining the format
    cv = Converter(pdf_path)
    cv.convert('temp.docx', start=0, end=None)
    cv.close()

    # Open the converted DOCX document
    doc = Document('temp.docx')

    # Configure Tesseract path (replace with your actual Tesseract executable path)
    pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

    content = []

    for paragraph in doc.paragraphs:
        paragraph_content = []
        for run in paragraph.runs:
            run_content = []
            for text, bold, italic in run._r.xml:
                if text:
                    run_content.append({"text": text, "bold": bold, "italic": italic})
            if run_content:
                paragraph_content.append({"type": "paragraph", "children": run_content})
        if paragraph_content:
            content.append(paragraph_content)

    # Extract tabular data, if any
    tables = []
    docx_text = docx2txt.process('temp.docx')
    for table in docx_text['tables']:
        table_content = []
        for row in table:
            row_content = []
            for cell in row:
                row_content.append({"text": cell})
            table_content.append({"type": "table-row", "children": row_content})
        tables.append({"type": "table", "children": table_content})

    # Remove the temporary DOCX file
    os.remove('temp.docx')

    # Combine both text and tables into the JSON content
    slate_js_json = [{"type": "document", "children": content + tables}]

    return json.dumps(slate_js_json, indent=4, ensure_ascii=False)

def save_json_to_file(json_content, output_file_path):
    with open(output_file_path, 'w', encoding='utf-8') as json_file:
        json_file.write(json_content)

# Specify the path for the input PDF
pdf_file_path = './Parikshitha_Hegde_Resume.pdf'

# Specify the path for the output JSON file
json_output_path = './output.json'

# Call the function to convert the PDF and perform OCR
slate_js_json = pdf_to_slatejs_json(pdf_file_path)

# Save the JSON content to a file
save_json_to_file(slate_js_json, json_output_path)
```
